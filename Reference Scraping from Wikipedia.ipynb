```python
import requests
from bs4 import BeautifulSoup

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}

url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

references = soup.find_all("div", class_="reflist")
ref_list = []
for reference in references:
    items = reference.find_all("li")
    count = 0
    for item in items:
        count = count + 1
        ref_list.append(item.text.replace("^"," ").strip())
```

```python
#code to get the heading
data = soup.find("div",id = "mw-content-text")
ref = data.find_all("h2")
heading= ""
for i in ref:
    if i.text == 'References':
        heading = i.text.upper()
```
